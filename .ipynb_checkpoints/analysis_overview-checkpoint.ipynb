# Responsible AI Results Overview

This notebook summarises the results of our fairness and explainability experiments for the **Adult Income** and **Financial Loan Access** datasets. We trained **Logistic Regression** and **Random Forest** models and evaluated their performance metrics (accuracy, precision, recall, F1, ROC-AUC) alongside group fairness metrics (Statistical Parity Difference, Equal Opportunity Difference, Equalised Odds Difference, Predictive Parity Difference). We also visualise the confusion matrices and feature importance for each model.

Use this notebook as an interpretive companion to the code package. Run the notebook in the project root so that the relative paths to the `results/` folder resolve correctly.

```python
import pandas as pd
from IPython.display import display

# Load results
perf = pd.read_csv('results/performance_metrics.csv')
fair = pd.read_csv('results/fairness_metrics.csv')

# Display performance metrics
print('Performance Metrics:')
display(perf)

# Display fairness metrics
print('Fairness Metrics:')
display(fair)
```

## Summary of Findings

**Adult Income dataset:**

- *Logistic Regression:* Accuracy ~0.81. The model shows noticeable disparities across gender and race with statistical parity differences of ~0.31 (gender) and ~0.28 (race) and equal opportunity differences of ~0.15 (gender) and ~0.29 (race). This suggests that the model predicts high income more frequently for some groups than others.
- *Random Forest:* Accuracy ~0.86. Fairness metrics improve somewhat, with gender SPD ~0.18 and race SPD ~0.17, and lower equal opportunity difference for gender (~0.09) though race EOD remains moderate (~0.27). This indicates that random forests perform better while being slightly fairer on gender but still show racial disparities.

**Financial Loan Access dataset:**

- *Logistic Regression:* Accuracy ~0.62. The model displays large disparities across race (SPD ~0.47, equal opportunity difference ~0.51) and gender (SPD ~0.26, EOD ~0.19). This indicates significant inequity in credit approval predictions between groups.
- *Random Forest:* Accuracy ~0.63. The model reduces disparity for race (SPD ~0.20, EOD ~0.21) and gender (SPD ~0.25, EOD ~0.15) while maintaining similar performance. However, fairness concerns remain, especially regarding race. Further mitigation strategies (re-weighting, threshold adjustment) would be advisable.

Overall, **Random Forest** generally delivers better accuracy and somewhat improved fairness compared with Logistic Regression. Nevertheless, persistent disparities—particularly for race—highlight the importance of applying Responsible AI techniques such as fairness-aware training, threshold tuning, or post-processing to reduce bias. These findings should be interpreted in the context of the dataset and not used to determine individual creditworthiness. The aim is to illustrate how model choices affect fairness metrics rather than to make high-stakes decisions.
